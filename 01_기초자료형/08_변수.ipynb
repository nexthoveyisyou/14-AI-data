```python
# 변수는 어떻게 만들까?
# 변수이름 = 변수에 저장할 값

# 변수명 규칙
# 1. 영문자, 숫자, 언더스코어(_) 만 사용 가능
# 2. 숫자로는 시작할 수 없음
# 3. 예약어(명령어)는 변수로 사용할 수 없다.
# 4. 대소문자 구분 한다.

name = "홍길동"
# 1name = "유재석" (x)
name2 = "강호동"

# user-name = "홍길동" (x)
user_name = "홍길동"

# 컨벤션
# user_name : 스네이크 스타일 (O) -> 가장 많이 씀
# UserName : 파스칼 스타일 (O) - 가장 많이 씀
# userName : 카멜 스타일 (O)

a = [1, 2, 3]
print(id(a))

b = [1, 2, 3]
print(id(b))

a.append(4)
print(id(a))
```

```python
b = a
print(id(b))
```

```python
a = [1, 2, 3]
b = a[:] # 복제 개념 (새롭게 생성됨)

print(id(b))
print(id(a))
```

```python
# 변수를 만드는 여러가지 방법
a ,b = ["python" , "like"]
print(a)
print(b)

print(type(a))
```

```python
(a ,b) = ["python" , "like"]
print(a)
print(b)

print(type(a))
```

```python
a = b = "python"
print(a)
print(b)
print(type(a))
```
