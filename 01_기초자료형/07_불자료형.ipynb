```python
print(1 == 1)
print(2 > 1)
print(1 > 2)
print(1 != 2)
```

```python
print(1, bool(1)) # 인수가 참인지 아닌지 구분
print(0, bool(0)) 
print(2, bool(2)) 
print("python", bool("python")) 
print("", bool("")) 
print("[]", bool([])) 
print("[1, 2, 3]", bool([1, 2, 3])) 
```

```python
if [1, 2, 3]:
    print("참")
else:
    print("거짓")
```

```python
# 논리 연산자
# 1. and 연산자

print(True and True )
print(False and True )
print(False and False )
print(True and True and False )

# 2. or 연산자
 
print(True or True )
print(False or True )
print(False or False )
print(True or True or False )

# 3. not 연산자
print(not True)
print(not False)
print(not 1)
print(not [])
```
